```python
# Construct spline
f_gamma = frini.get_splines()["base"]
```

```python
test_egrid = np.logspace(-12,-7,100)
for zi, z in enumerate(frini.z_dist):
    plt.loglog(test_egrid,5**zi*test_egrid*f_gamma(test_egrid,z),ls=':',color='k')
    plt.loglog(frini.energy[zi],5**zi*frini.energy[zi]*frini.ph_density[zi])
plt.ylim(1e-4,1e8)
```

```python
from scipy.integrate import simps
print('Power integrals')
print('original table, orig. E-grid, arb. E-grid, 2D interp.')
for i, z in enumerate(frini.z_dist): 
     
    orig = simps(frini.energy[i]*frini.ph_density[i], frini.energy[i])
    ogri = simps(frini.energy[i]*frini.z_n_splines[i](frini.z_e_splines[i](frini.energy[i])),frini.energy[i])
    agri = simps(frini.common_evec*frini.z_n_splines[i](frini.z_e_splines[i](frini.common_evec)),frini.common_evec)
    fspl = simps(test_egrid*f_gamma(test_egrid,z),test_egrid)
    print(('z={z}: {0:4.3e} / {1:4.3e} / {2:4.3e} / {3:4.3e}, ' + 
        'rel. err.: {4:4.3f} / {5:4.3f} / {6:4.3f}').format(
        orig, ogri, agri, fspl, ogri/orig, agri/orig, fspl/orig, z=frini.z_dist[i]))
```

# Inoue EBL

4) CIB spectrum by Inoue et al.

Data files are taken from CRPropagation5 (based on paper)

```python
inoue = ppf.Inoue2013()
ispl = inoue.get_splines()

inoue_f_gamma = ispl["base"]
inoue_f_gamma_min = ispl["lower"]
inoue_f_gamma_max = ispl["upper"]
```

```python
import matplotlib
#Set up color scheme
norm = matplotlib.colors.Normalize(
    vmin=0,vmax=10)
c_m = matplotlib.cm.RdBu
s_m = matplotlib.cm.ScalarMappable(cmap=c_m, norm=norm)
s_m.set_array([])

test_egrid = np.logspace(-12,-7,100)

fig, ax = plt.subplots(figsize=(8,6))

for zi in np.arange(0,110,10):
    z = inoue.z_dist[zi]
    sc = 10**(zi/10)
    if z > 0.1:
        zi = zi-1
        z = inoue.z_dist[zi]
        
    
    inspec = sc*inoue.energy*inoue.ph_density_base[zi,:]
    plt.loglog(inoue.energy, inspec,
               color=s_m.to_rgba(10.-z),lw=2.,
              label = 'Inoue et al.' if zi == 0 else '_nolabel_')
    
    plt.loglog(test_egrid,sc*test_egrid*inoue_f_gamma(test_egrid,z,),
               ls=':',color='k',
              label = '2D spline (Inoue)' if zi == 0 else '_nolabel_')
    plt.loglog(test_egrid,sc*test_egrid*f_gamma(test_egrid,z),
               ls='-.',lw=2.,color=s_m.to_rgba(10.-z),
               label = '2D spline (Franceschini)' if zi == 0 else '_nolabel_')
    plt.text(2.5e-13 if z < 10 else 5e-13,inspec[0],str(z))
plt.text(1.5e-13,1e9,'Cosmic infrared background models')
plt.text(2.2e-13,2e6,'Redshift')
plt.ylim(1e-5,1e10)
plt.xlim(1e-13,1e-7)
plt.xlabel(r'Photon energy $\epsilon$ (GeV)')
plt.ylabel(r' $10^{z_i} \epsilon$ d$n$/d$\epsilon$ (cm$^{-3})$')
plt.legend(loc='upper right',fontsize='small',frameon=False)
plt.tight_layout()
plt.savefig('CIB_models.png',dpi=300)
```
